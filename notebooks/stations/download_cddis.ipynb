# Download DORIS STCD Data from CDDIS

This notebook downloads DORIS STCD products from the CDDIS archive using the implemented Python library.

## Imports and project setup

**Standard library**
- `sys` – Python path handling
- `pathlib.Path` – filesystem paths

**Project library**
- `doris.input.cddis.download_from_cddis` – high-level CDDIS download workflow

```python
# --- Standard library ---
import sys                                # system-specific parameters and functions
from pathlib import Path                  # filesystem paths


# --- doris packages ---
_src = Path("../../src").resolve()
if str(_src) not in sys.path:
    sys.path.insert(0, str(_src))

try:
    from doris.input.cddis import download_from_cddis

except ImportError as e:
    raise RuntimeError(
        "Missing doris modules. Make sure the project src path is on PYTHONPATH "
        "or run the notebook from the expected directory: notebooks/stations."
    ) from e
```

## Import the CDDIS workflow

The notebook uses the high-level `download_from_cddis` function from the local `doris` package.

```python
# --- Dataset configuration ---
TECHNIQUE = "doris".strip().lower()
SUBTREE   = "products".strip().lower()
PRODUCT   = "stcd".strip().lower()
SOLUTION  = "gop25wd04".strip().lower()
SATELLITE = None


# --- Paths ---
OUTPUT_ROOT = Path("../../data")
LOGIN_FILE  = Path("login.txt")


# --- Print configuration ---
print("Technique:", TECHNIQUE)
print("Subtree:", SUBTREE)
print("Product:", PRODUCT)
print("Solution:", SOLUTION)
print("Satellite:", SATELLITE)

print("\nOutput root:", OUTPUT_ROOT.resolve())
print("Login file:", LOGIN_FILE.resolve())
print("Login file exists:", LOGIN_FILE.exists())
```

## Run CDDIS download workflow

This step executes the high-level download workflow.

The workflow performs:
- authentication to the CDDIS archive,
- retrieval and parsing of the `MD5SUMS` index,
- download of all matching archive files,
- optional decompression of downloaded files.

Parallel download is enabled to improve performance.

```python
# --- Download options ---
OVERWRITE = True
PARALLEL_DOWNLOAD = True
MAX_WORKERS = 16

DECOMPRESS = True
KEEP_COMPRESSED = False
OVERWRITE_DECOMPRESSED = True


# --- Run CDDIS workflow ---
result = download_from_cddis(
    technique=TECHNIQUE,
    subtree=SUBTREE,
    product=PRODUCT,
    solution=SOLUTION,
    satellite=SATELLITE,
    output_root=OUTPUT_ROOT,
    login_file=LOGIN_FILE,
    overwrite=OVERWRITE,
    parallel_download=PARALLEL_DOWNLOAD,
    max_workers=MAX_WORKERS,
    decompress=DECOMPRESS,
    keep_compressed=KEEP_COMPRESSED,
    overwrite_decompressed=OVERWRITE_DECOMPRESSED)
```

## Download summary

```python
print("Dataset URL:", result.dataset_index.dataset_url)
print("MD5SUMS:", result.dataset_index.md5_path)
print("Found archives:", len(result.dataset_index.archive_names))
print("Downloaded files:", len(result.downloaded_paths))
print("Decompressed files:", len(result.decompressed_paths))
```
