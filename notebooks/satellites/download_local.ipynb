## Imports

```python
# --- doris packages ---
_src = Path("../../src").resolve()
if str(_src) not in sys.path:
    sys.path.insert(0, str(_src))

from doris.input.local import copy_from_local
```

## Paths and parameters

```python
from pathlib import Path

source_dir = Path(r"../../data\amalie_sp3_2024\gop")
local_dir = Path("../../data") / "orbits"
solution = "gop"
satellite = "srl"
filename_pattern = None

source_dir = source_dir/satellite

# Pro kopii vsech souboru:
# solution = None
# satellite = None

print("Source dir:", source_dir)
print("Local dir:", local_dir)
print("Solution filter:", solution)
print("Satellite filter:", satellite)
print("Filename pattern:", filename_pattern)
```

## Copy

```python
result = copy_from_local(
    source_dir=source_dir,
    local_dir=local_dir,
    filename_pattern=filename_pattern,
    solution=solution,
    satellite=satellite,
    overwrite=True,
    decompress=False,
    keep_compressed=False,
    overwrite_decompressed=True,
)
```

## Summary

```python
print("Source dir:", source_dir.resolve())
print("Target dir:", local_dir.resolve())
print("Found files:", result.count)
print("Copied files:", result.count_copied)
print("Skipped existing:", result.count_skipped_existing)
print("Decompressed files:", result.count_decompressed)
```
