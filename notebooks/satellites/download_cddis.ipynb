## Imports

```python
# --- Standard library ---
import sys                                # system-specific parameters and functions
from pathlib import Path                  # filesystem paths


# --- doris packages ---
_src = Path("../../src").resolve()
if str(_src) not in sys.path:
    sys.path.insert(0, str(_src))

try:
    from doris.input.cddis import download_from_cddis

except ImportError as e:
    raise RuntimeError(
        "Missing doris modules. Make sure the project src path is on PYTHONPATH "
        "or run the notebook from the expected directory: notebooks/stations."
    ) from e
```

## Paths and parameters

```python
TECHNIQUE = "doris"
SUBTREE = "PRODUCTs"
PRODUCT = "orbits"
SOLUTION = "ssa"
SATELLITE = "srl"

output_root = Path("../../data")
login_file = Path("login.txt")

print("TECHNIQUE:", TECHNIQUE)
print("SUBTREE:", SUBTREE)
print("PRODUCT:", PRODUCT)
print("SOLUTION:", SOLUTION)
print("SATELLITE:", SATELLITE)
print("Output root:", output_root)
print("Login file:", login_file.resolve())
print("Login file exists:", login_file.exists())
```

## Download

```python
result = download_from_cddis(
    technique=TECHNIQUE,
    subtree=SUBTREE,
    product=PRODUCT,
    solution=SOLUTION,
    satellite=SATELLITE,
    output_root=output_root,
    login_file=login_file,
    overwrite=True,
    parallel_download=True,
    max_workers=16,
    decompress=True,
    keep_compressed=False,
    overwrite_decompressed=True,
)
```

## Summary

```python
print("Dataset URL:", result.dataset_index.dataset_url)
print("MD5SUMS:", result.dataset_index.md5_path)
print("Found archives:", len(result.dataset_index.archive_names))
print("Downloaded files:", len(result.downloaded_paths))
print("Decompressed files:", len(result.decompressed_paths))
```
